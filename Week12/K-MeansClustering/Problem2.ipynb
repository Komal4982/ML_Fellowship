# Step4:Getting the input data

```python
# get input data as x
x=dataset.iloc[:10000,:].values
x.shape
```

# Step5:Using the elbow method to find the optimal no of clusters

```python
#using the elbow method to find the optimal no of clusters
import sklearn
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# within cluster sum of square(wcss)
wcss=[]
for i in range(1,20):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
#     print(kmeans)
    kmeans.fit(x)
#     print(kmeans)
    wcss.append(kmeans.inertia_)
# print("WCSS",wcss)
plt.plot(range(1,20),wcss)
plt.title("the elbow method")
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
    
```

# Step6:applying k-means to the dataset

```python
#applying k-means to the dataset
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
# print(kmeans)
y_means=kmeans.fit_predict(x)
y_means.shape
```
