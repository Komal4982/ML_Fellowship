```python
class Linear_regression2:
    def __init__(self):
        
        self.learning_rate = 0.01
        self.theta_0 = 0.5
        self.theta_1 = 0.5
        self.epoch = 1000
        
    
    #missing data
    def missing_data(self):
        df.replace(np.NaN,df.mean(),inplace=True)
        return df
    
  
    def remove_outliers(self,df):
        low=.05
        high=.95
        quantile_df=df.quantile([low,high])
        for name in list (df.columns):
            if ptypes.is_numeric_dtype(df[name]):
                df=df[(df[name]>quantile_df.loc[low,name])&(df[name]<quantile_df.loc[high,name])]
        return df
    
    def feature_scaling(self,feature_data):
        for data in feature_data.columns:
            feature_data[data]=(feature_data[data]-feature_data[data].mean())/np.std(feature_data[data])
            return feature_data
        
    
    def gradient_decent(self,x_data,y_data):
        theta_0=self.theta_0
        theta_1=self.theta_1
        for iteration in range(self.epoch):
            cost=0
            h1=h2=0.0
            cost_temp=0.0
            
            for data in range(len(x_data)):
                hypothesis =theta_0 +(theta_1 * x_data[data])
                h1 +=(hypothesis-y_data[data])
                h2 += ((hypothesis-y_data[data])*x_data[data])
                
                cost += hypothesis-y_data[data]
                
             # getting minimum cost function basic
            cost_temp += (hypothesis-y_data[data])**2
                      
            cost=(1/2*len(x_data))*cost_temp
                      
            theta_0=theta_0-((self.learning_rate*h1)/len(x_data))
            
            theta_1=theta_1-((self.learning_rate*h2)/len(x_data))
        return theta_0,theta_1
        
        
    
    def predict_values(self,y_data,theta_0,theta_1):
        y_predict_values=[0]*len(y_data)
        for i in range(len(y_data)):
            y_predict_values[i]=(theta_0+(theta_1*y_data[i]))
        return y_predict_values
    
    def accuracy(self,y_data,y_predict_data):
        diff=0
        for i in range(len(y_data)):
            diff +=abs(( y_predict_data[i]-y_data[i])/y_data[i])
        diff=(diff/len(y_data))
        data_accuracy=1-diff
        return data_accuracy*100

  
        
        
        
        
        
def main():
    obj=Linear_regression2()
    
    df=obj.missing_data()
    #print(df)
    
    remove_outliers=obj.remove_outliers(df)
    #print("remove outliers""\n",remove_outliers)
    
    feature_data=obj.feature_scaling(remove_outliers)
    #print("featured data""\n",feature_data)
    
    x_data=np.array(df['x'])
    #print("seperate data list into array:=""\n",x_data)
    
    y_data=np.array(df['y'])
    #print("seperate data list into array:=""\n",y_data)
    
    theta_0,theta_1=obj.gradient_decent(x_data,y_data)
    print("gradient decent for theta_0",theta_0)
    print("gradient decent for theta_1",theta_1)
    
    
    y_predict_data=obj.predict_values(y_data,theta_0,theta_1)
    #print(y_predict_data)
    
    accuracy=obj.accuracy(y_data,y_predict_data)
    print(accuracy)
    
    
    
    

if __name__=='__main__':
    main()
```
